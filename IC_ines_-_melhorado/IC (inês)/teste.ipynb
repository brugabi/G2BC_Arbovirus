# Preparar o ambiente

```python
import pandas as pd
from scripts.webscrapping import extract_pcmid
```

# Extract

```python
# Importando o csv
data = pd.read_csv('data/raw/DENV2_locus_pubmed.csv')
```

# Transform

```python
# Coluna Pubmed Ã© uma string em forma de lista
def transform_string_in_list(string):
    string = string.replace('[','').replace(']','').replace("'",'')
    lista = string.split(',')
    return lista
data['Pubmed accession number'] = data['Pubmed accession number'].apply(transform_string_in_list)
```

```python
# Explodindo o pubmed
data_explouded = data.explode('Pubmed accession number')
```

```python
# Dropando as linhas duplicadas
data_explouded.drop_duplicates(inplace=True)
```

```python
# Aplicando strip
data_explouded['Pubmed accession number'] = data_explouded['Pubmed accession number'].apply(str.strip)
```

```python
# Pegando a lista de pubmeds
lista_pubmed = set(data_explouded['Pubmed accession number'])
```

```python
# Extraindo os pmcid
lista_com_os_pmcid = extract_pcmid(pubmed_list=lista_pubmed)
```

```python
# Inserindo o PMCID do respectivo pubmed
def insert_pmcid(pubmed,lista_pmcid):
    for tupla in lista_pmcid:
        if pubmed == tupla[0].strip():
            return tupla[1]

data_explouded['pmcid'] = data_explouded['Pubmed accession number'].apply(lambda x: insert_pmcid(pubmed=x,lista_pmcid=lista_com_os_pmcid))
   
```

# Load

```python
data_explouded.to_csv('data/processed/data_processed.csv',index=False)
```
